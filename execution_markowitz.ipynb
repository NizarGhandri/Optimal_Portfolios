### Efficient Frontier Plot

```python
def plot_efficient_frontier(portfolio_df):
    # find minimum volatility and maximum sharpe values in the portfolio
    min_volatility = portfolio_df['Volatility'].min()
    max_sharpe = portfolio_df['Sharpe Ratio'].max()
    
    # min-variance portfolio
    min_variance_portfolio = portfolio_df.loc[portfolio_df['Volatility'] == min_volatility]
    # sharpe portfolio
    sharpe_portfolio = portfolio_df.loc[portfolio_df['Sharpe Ratio'] == max_sharpe]
    
    # plot highliting the optimal portfolio and minimum volatility portfolio
    plt.style.use('seaborn-dark')
    portfolio_df.plot.scatter(x='Volatility', 
                              y='Returns', 
                              c='Sharpe Ratio',
                              cmap='RdYlGn', 
                              edgecolors='black', 
                              figsize=(10, 8), 
                              grid=True)
    plt.scatter(x=sharpe_portfolio['Volatility'], 
                y=sharpe_portfolio['Returns'], 
                c='red', 
                marker='D', 
                s=200)
    plt.scatter(x=min_variance_portfolio['Volatility'], 
                y=min_variance_portfolio['Returns'], 
                c='blue', 
                marker='D', 
                s=200 )
    plt.xlabel('Volatility (Std. Deviation)')
    plt.ylabel('Expected Returns')
    plt.title('Efficient Frontier')
    plt.show()
```

```python
plot_efficient_frontier(markowitz[2])
```
